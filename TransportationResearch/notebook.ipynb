```
mothly_transportation_data= pd.read_csv("/work/Monthly_Transportation_Statistics.csv")
```

```
df = mothly_transportation_data[["Date",'U.S. Airline Traffic - International - Seasonally Adjusted', 'U.S. Airline Traffic - Domestic - Seasonally Adjusted', 'Highway Vehicle Miles Traveled - All Systems',  'Transit Ridership - Fixed Route Bus - Adjusted','Transit Ridership - Urban Rail - Adjusted','Transit Ridership - Other Transit Modes - Adjusted']]
df_rename = df.rename(columns={"U.S. Airline Traffic - International - Seasonally Adjusted": "International Airline Traffic","U.S. Airline Traffic - Domestic - Seasonally Adjusted": "Domestic Airline Traffic","B": "Highway Vehicle Miles Traveled","Highway Vehicle Miles Traveled - All Systems": "Highway Vehicle Miles Traveled","Transit Ridership - Fixed Route Bus - Adjusted": "Bus Ridership","Transit Ridership - Urban Rail - Adjusted": "Rail Ridership","Transit Ridership - Other Transit Modes - Adjusted": "Other Transit Ridership"})
dftest = df_rename.copy()
dfdropped = dftest.dropna()
dfdropped['Date'] =  pd.to_datetime(dfdropped['Date']).dt.date
dfdropped = dfdropped.reset_index(drop=True)
label_list = []
for i in range(26):
    label_list.append(0)
for i in range(22):
    label_list.append(1)
len(label_list)
dfdropped["treatment_label"] = label_list
dfq1 = dfdropped.copy()
dfq1
```

```
def difference_in_means(df, binary_col, numerical_col):
    mean = df.groupby(binary_col).agg(np.mean).reset_index()
    diff = mean[numerical_col][1] - mean[numerical_col][0]
    return diff

def one_simulated_difference(df, binary_col, numerical_col):
    shuffled_labels = df[binary_col].sample(frac=1, replace=False).values
    df['Shuffled Label'] = shuffled_labels
    diff = difference_in_means(df, 'Shuffled Label', numerical_col)
    return diff
    
def avg_difference_in_means(df, binary_col, numerical_col):
    selected_df = df[[binary_col, numerical_col]]
    original_abs_diff = difference_in_means(selected_df, binary_col, numerical_col) 
    
    differences = []
    repetitions = 1000
    for i in np.arange(repetitions):
        new_difference = one_simulated_difference(selected_df, binary_col, numerical_col)    
        differences = np.append(differences, new_difference) 

    p_value = np.count_nonzero(original_abs_diff >= differences)/repetitions
    return p_value
```

```
p_value1 = avg_difference_in_means(dfq1, 'treatment_label', 'International Airline Traffic')
p_value1
```

```
p_value2 = avg_difference_in_means(dfq1, 'treatment_label', 'Domestic Airline Traffic')
p_value2
```

```
p_value3 = avg_difference_in_means(dfq1, 'treatment_label', 'Highway Vehicle Miles Traveled')
p_value3
```

```
p_value4 = avg_difference_in_means(dfq1, 'treatment_label', 'Other Transit Ridership')
p_value4
```

```
p_value5 = avg_difference_in_means(dfq1, 'treatment_label', 'Rail Ridership')
p_value5
```

```
p_value6 = avg_difference_in_means(dfq1, 'treatment_label', 'Bus Ridership')
p_value6
```

```
p_values = [p_value1, p_value2 ,p_value3 ,p_value4 ,p_value5 ,p_value6]
def naive(alpha):
    decisions = (alpha) >= np.array(p_values)
    return decisions
naive(0.05)
```

```
# Benjamin Hochberg threshold
def modified_benjamini_hochberg(p_values, alpha):
    sorted_pval = np.sort(p_values)
    fail = []
    for i in reversed(range(len(p_values))):
        if sorted_pval[i] <= (i+1)*(alpha/len(p_values)):
            break;
        else:
            fail += [sorted_pval[i]]

    decisions = p_values <= sorted_pval[i]
    return decisions, fail, sorted_pval[i]

p_values = [p_value1, p_value2 ,p_value3 ,p_value4 ,p_value5 ,p_value6]
decidions, fail_p_val, boundary_p = modified_benjamini_hochberg(p_values, 0.05)
decidions
```

Under the Benjamin Hochberg threshold, we reject all the null hypothesis since all the p-values are smaller than less than or equal to k*alpha/n. It suggests that international airline traffic, domestic airline traffic, highway vehicle, rail ridership, bus ridership, and other ridership all decreased after the pandemic. Comapre to the naive threshold, this method controls the False Discovery Rate.

```
# Bonferroni correction

def bonferroni(p_values, alpha_total): 
    decisions = (alpha_total/len(p_values)) >= np.array(p_values)
    return decisions

bonferroni(p_values, 0.05)
```

Using Bonferroni Correction, we want to control the Family Wise Error Rate to be 0.05, so we threshold all the p-values at 0.05/6. In this case, we fail to reject only one hypothesis. That is, there is not significant evidence to conclude that the highway vehicle miles traveled decreased after the pandemic. This was implyed in the EDA. We saw highway vehicle miles traveled decreased sharply in 2020 march and rebounced back to it original state in 6 months.

```
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm
import itertools
from ipywidgets import interact, interactive
from scipy import stats

sns.set(style="dark")
plt.style.use("ggplot")
%matplotlib inline
```

```
mothly_transportation_data= pd.read_csv("/work/Monthly_Transportation_Statistics.csv")
mothly_transportation_data
```

```
dfq2 = mothly_transportation_data[["Date",
'State and Local Government Construction Spending - Mass Transit',
'Transit Ridership - Other Transit Modes - Adjusted',
'Transit Ridership - Fixed Route Bus - Adjusted',
'Transit Ridership - Urban Rail - Adjusted',
'State and Local Government Construction Spending - Highway and Street',
'Real Gross Domestic Product - Seasonally Adjusted']]
dfdropped_q2 = dfq2.dropna()
dfdropped_q2['Date'] =  pd.to_datetime(dfdropped_q2['Date']).dt.date
dfdropped_q2 = dfdropped_q2.reset_index(drop=True)
renamed_df = dfdropped_q2
renamed_df = renamed_df.rename(columns={'State and Local Government Construction Spending - Mass Transit':'Transit Spending', 
'Transit Ridership - Other Transit Modes - Adjusted': 'Other Transit', 
'Transit Ridership - Fixed Route Bus - Adjusted': 'Bus', 
'Transit Ridership - Urban Rail - Adjusted': 'Rail',
'State and Local Government Construction Spending - Highway and Street': 'Highway Spending',
'Real Gross Domestic Product - Seasonally Adjusted': 'Real GDP'})
renamed_df[['Total Ridership']] = renamed_df['Other Transit'].values + renamed_df['Bus'].values + renamed_df['Rail'].values 
renamed_df 
```

```
fig = sns.pairplot(renamed_df,  plot_kws=dict(marker="o", alpha = 0.5))
for i, j in zip(*np.triu_indices_from(fig.axes, 1)):
    fig.axes[i, j].set_visible(False)
plt.show()
```

Z = treatment = State and Local Government Construction Spending - Mass Transit

W = instrumental = State and Local Government Construction Spending - Highway and Street

Y = outcome = Transit Ridership - Other Transit Modes - Adjusted + Transit Ridership - Fixed Route Bus - Adjusted + Transit Ridership - Urban Rail - Adjusted

X = confounder = Real Gross Domestic Product - Seasonally Adjusted

```
#Function to fit OLS model
def fit_OLS_model(df, target_variable, explanatory_variables, intercept = False):
    """
    Fits an OLS model from data.
    
    Inputs:
        df: pandas DataFrame
        target_variable: string, name of the target variable
        explanatory_variables: list of strings, names of the explanatory variables
        intercept: bool, if True add intercept term
    Outputs:
        fitted_model: model containing OLS regression results
    """
    
    target = df[target_variable]
    inputs = df[explanatory_variables]
    if intercept:
        inputs = sm.add_constant(inputs)
    
    fitted_model = sm.OLS(target, inputs).fit()
    return(fitted_model)
```

```
# Fit OLS parameters to predict treatment Z using instructmental W
gamma1_model = fit_OLS_model(renamed_df, 'Transit Spending', ['Highway Spending'], intercept = True)
print(gamma1_model.summary())
```

```
# Compute predictions for total ridership
intercept_OLS = gamma1_model.params[0]
gamma1_OLS = gamma1_model.params[1]
Z_hat = intercept_OLS + gamma1_OLS*renamed_df['Highway Spending']

# Add the treatment predictions to the trimmed_df dataframe
renamed_df['PredictedTransitSpending'] = Z_hat
renamed_df.head()
```

```
# Fit OLS parameters to predict outcome Y using the predicted Z_hat.
beta1_model = fit_OLS_model(renamed_df, 'Total Ridership', ['PredictedTransitSpending'], intercept = True)
print(beta1_model.summary())
```

```
plt.hist(renamed_df['PredictedTransitSpending'], alpha=0.5)
plt.hist(renamed_df['Transit Spending'], alpha=0.5)

plt.legend(['Z_hat','Z'])
plt.show()
```

```
# Using confounder X and treatment Z to regress on Y (1 stage OLS)
beta2_model = fit_OLS_model(renamed_df, 'Total Ridership', ['Real GDP', 'Transit Spending'], intercept = True)
print(beta2_model.summary())
```
